## Assignment 1.5 (4 points)
Write a function that takes a product review as input (as a string) and determines whether the review is positive, negative, or neutral based on specific keywords. The function should return one of three strings: 'Negative review', 'Positive review', or 'Neutral review'.

The function should consider words like 'excellent', 'good', 'amazing' as positive, and words like 'poor', 'bad', 'terrible' as negative. If a review contains both positive and negative keywords, or if it contains neither, it should be classified as neutral.

For example, the following function calls should return 'Negative review':

`reviewSentiment('The quality of this product was poor and I'm disappointed.')`

`reviewSentiment('Terrible experience, the product broke on the first day.')`

The following function calls should return 'Positive review':

`reviewSentiment('Absolutely love this product, it's amazing!')`

`reviewSentiment('Good quality and excellent value for money.')`

The following function calls should return 'Neutral review':

`reviewSentiment('This product has both good and bad features.')`

`reviewSentiment('I haven't decided if I like this product.')`

Note that your function should work on other reviews as well.

`````python
# ANSWER CELL 1.5 (DO NOT REMOVE OR CHANGE THIS LINE)

def reviewSentiment(review):
    # pass (Here as well, I was not sure about deleteing pass so I just commented out.)
    my_dictionary = {"great": 2 , "fantastic": 2, "excellent": 2, "amazing": 2, "love": 2, "enjoyed": 1, "good": 1, "awful": -2, "terrible": -2, "disappointing": -2, "worst": -2, "poor": -1, "bad": -1,}

    score = 0
    words = review.split() # use split to split the string into a list of words

    # This punctuation list is from ChatGPT
    punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'

    for word in words:
        # Here checks each character char in the word, and includes the character in the new list
        # only if it is not in punctuation (the defined string of punctuation characters).
        fixed_word = ''.join(char for char in word if char not in punctuation).lower() #  Joins all its elements together to form a new string and convert it to lowercase
        if fixed_word in my_dictionary:
            score += my_dictionary[fixed_word]
    
    if score > 0:
        return "Positive review"
    elif score < 0:
        return "Negative review"
    else:
        return "Neutral review"

# Test
print(reviewSentiment('The quality of this product was poor and I am disappointed.'))
print(reviewSentiment('Absolutely love this product, it is amazing!')) # Here I see code manages to work even if I delete "love" since I removed the punctuation (amazing!)
print(reviewSentiment('I have not decided if I like this product.'))
print(reviewSentiment('Good quality and excellent value for money.'))
print(reviewSentiment('Terrible experience, the product broke on the first day.'))
print(reviewSentiment('I thought it was good but it was actually awful.'))
print(reviewSentiment('It is good and bad. I cannot decide yet.'))
`````

Remember, regular expressions use special characters to denote different types of patterns. For example, `\d` matches any digit, and `\b` denotes a word boundary, which is useful for extracting whole words or numbers.

Test your regular expressions as you build them to ensure they match the desired patterns correctly. Python's `re.findall()` function can be used to find all occurrences of the pattern in the given text.

For additional details, check out: https://docs.python.org/3/library/re.html

`````python
# ANSWER CELL 1.6 (DO NOT REMOVE OR CHANGE THIS LINE)

import re

text = """
  John Doe, Email: john.doe@example.com, Phone: 96 23 45 67
  Jane Smith, Email: jane_smith@example.net, Phone: 91 56 78 90
  Alex Johnson, Email: alex.j@example.org, Phone: 92 34 56 78
"""
# Chiaki Nishihara, Email: chiaki@example.no, Phone: 92 98 89 05 (Tested)

email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b' # b means word boundary, and here is the structure
emails = re.findall(email_pattern, text)
print(emails)

phone_pattern = r'\d{2} \d{2} \d{2} \d{2}' # d means digit. Here we talk about Norwegian phone number type.
phone_numbers = re.findall(phone_pattern, text)
print(phone_numbers)

name_pattern = r'\b[A-Z][a-z]* [A-Z][a-z]*\b'
# List comprehension
names_with_e = [name for name in re.findall(name_pattern, text) if 'e' in name] # Write in a list and use for loop to check if there is "e" in the name
print(names_with_e)
`````

`````python
# ANSWER CELL 2.4.2 (DO NOT REMOVE OR CHANGE THIS LINE)
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 3],
              [2, -5]])

B = np.array([[20,],
              [-5]])

solution = np.linalg.solve(A, B)

x_values = np.linspace(-10, 10)

y_values_equation1 = (20 - 4 * x_values) / 3
y_values_equation2 = (2 * x_values + 5) / 5

x_solution = solution[0, 0] # From the solution
y_solution = solution[1, 0] # Here too, from the solution


plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values_equation1, label='4x + 3y = 20')
plt.plot(x_values, y_values_equation2, label='2x - 5y = -5')

plt.plot(x_solution, y_solution, 'g*' ,markersize = 10) # Here I use green star
plt.text(x_solution, y_solution, f'({x_solution:.2f}, {y_solution:.2f})') # Here print the solution, 2f means 2 decimal places

plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5) # axhline() function adds a horizontal line across the axis
plt.axvline(0, color='black',linewidth=0.5) # axvline() function adds a vertical line across the axis
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5) # I wanted grid line to be dashed line so chose "--".
plt.legend()
plt.title('Graphs of 4x + 3y = 20 and 2x - 5y = -5')
plt.show()
`````

4. (3 points) Create a query that returns all the information from the customer table for all the customers who purchased a Coupe car for more than 150,000.

`````python
%%sql --ANSWER CELL 4.1.1 (DO NOT REMOVE OR CHANGE THIS LINE)
SELECT * FROM Employees WHERE department = 'Service';
`````

`````python
%%sql --ANSWER CELL 4.1.2 (DO NOT REMOVE OR CHANGE THIS LINE)
SELECT * FROM Cars WHERE year = 2015 LIMIT 15;
`````

`````python
%%sql --ANSWER CELL 4.1.3 (DO NOT REMOVE OR CHANGE THIS LINE)

SELECT Sales.date, Sales.list_price, Employees.name, Cars.make
FROM Sales
JOIN Cars ON Sales.car_id = Cars.id
JOIN Employees ON Sales.employee_id = Employees.id
WHERE Cars.type = 'SUV' AND Sales.list_price > 170000;
`````

`````python
%%sql --ANSWER CELL 4.1.4 (DO NOT REMOVE OR CHANGE THIS LINE)

SELECT Customers.*
FROM Customers
JOIN Sales ON Customers.id = Sales.customer_id
JOIN Cars ON Sales.car_id = Cars.id
WHERE Cars.type = 'Coupe' AND Sales.price > 150000;
`````

`````python
# ANSWER CELL 4.2.1 (DO NOT REMOVE OR CHANGE THIS LINE)

import sqlite3
import pandas as pd

con = sqlite3.connect('auto_dealership_exam.db')

# I write like below because it looks easier to read
nissan_cars_query = """
    SELECT *
    FROM Cars
    WHERE make = 'Nissan'
"""
nissan_cars_df = pd.read_sql_query(nissan_cars_query, con)
con.close()

print("Pandas DataFrame with all the info about the Nissan cars:")
print(nissan_cars_df)
`````

`````python
# ANSWER CELL 4.2.2 (DO NOT REMOVE OR CHANGE THIS LINE)

con = sqlite3.connect('auto_dealership_exam.db')

ford_cars_query = """
    SELECT year, msrp
    FROM Cars
    WHERE make = 'Ford'
"""
ford_cars_df = pd.read_sql_query(ford_cars_query, con)

ford_yearly_mean = ford_cars_df.groupby('year')['msrp'].mean()

con.close()

plt.figure(figsize = (12, 6))
ford_yearly_mean.plot(kind = 'bar', color = 'red')
plt.title('Yearly Average MSRP of Ford Cars')
plt.xlabel('Year')
plt.ylabel('Mean MSRP')
plt.xticks(rotation = 48) # Rotation makes it easier to see the years!
plt.show()
`````

`````python
%%sql --ANSWER CELL 4.2.3 (DO NOT REMOVE OR CHANGE THIS LINE)

SELECT Employees.name, SUM(Sales.price) AS total_sales
FROM Sales
JOIN Employees ON Sales.employee_id = employee_id
GROUP BY Sales.employee_id
ORDER BY total_sales
LIMIT 1;
`````

`````python
%%sql --ANSWER CELL 4.2.4 (DO NOT REMOVE OR CHANGE THIS LINE)

SELECT make, year, msrp
FROM Cars
WHERE msrp > 3 * (SELECT AVG(msrp) FROM Cars);
`````
