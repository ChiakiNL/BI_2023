## Exercise 1: Basic Line Plot
Given the following data, plot the relationship between x and y.

x = [0, 1, 2, 3, 4, 5] 

y = [0, 1, 4, 9, 16, 25]

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create the plot
# ↓この行では、新しい図（Figure）を作成しています。figsize パラメータを用いて、図のサイズを指定しています（幅が 10 インチ、高さが 6 インチ）。
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='*', linestyle='-', color='r')

# Add titles and labels
# これらの行では、プロットにタイトルと軸ラベルを追加しています。title 関数は図のタイトルを設定し,
# xlabel と ylabel 関数はそれぞれ x 軸と y 軸のラベルを設定します。
plt.title('Relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

## Exercise 2: Scatter Plot
Create a scatter plot with the below data. Customize the plot by adding labels, a title, and changing the marker color to red.

x = [1, 3, 5, 7, 9] 

y = [10, 5, 8, 6, 7]

```python
import matplotlib.pyplot as plt

# Given data
x = [1, 3, 5, 7, 9]
y = [10, 5, 8, 6, 7]

# Create the scatter plot
# この行では、新しい図（Figure）を作成しています。figsize パラメータを用いて、図のサイズを指定しています（幅が 6インチ、高さが 4インチ）。
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='red', marker='^')

# Add titles and labels
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

## Exercise 3: Bar plot
Create a bar plot representing the sales of three products.

products = ['Product A', 'Product B', 'Product C']
sales = [100, 150, 80]

```python
import matplotlib.pyplot as plt

# Given data
products = ['Product A', 'Product B', 'Product C']
sales = [100, 150, 80]

# Create the bar plot
plt.figure(figsize=(6, 4))
plt.bar(products, sales, color='blue')

# Add titles and labels
plt.title('Sales of Three Products')
plt.xlabel('Products')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

## Exercise 4: Multiple Plots in One Figure
Plot two lines in the same figure to represent the stock prices of two companies over time. Add a legend to distinguish the companies.

days = [1, 2, 3, 4, 5] 

company_A_prices = [50, 51, 52.5, 50.5, 51.5] 

company_B_prices = [55, 53, 54, 54.5, 53.5]

```python
import matplotlib.pyplot as plt

# Given data
days = [1, 2, 3, 4, 5]
company_A_prices = [50, 51, 52.5, 50.5, 51.5]
company_B_prices = [55, 53, 54, 54.5, 53.5]

# Create the plot
plt.figure(figsize=(8, 5))

# Plot Company A stock prices
# x, yに合わせてdays, company_A_pricesの順にあてはめる
plt.plot(days, company_A_prices, label='Company A', marker='o', linestyle='-', color='blue')

# Plot Company B stock prices
plt.plot(days, company_B_prices, label='Company B', marker='x', linestyle='--', color='red')

# Add titles, labels, and legend
# plt.legend()関数は、グラフ上で各データ系列（線、点、棒など）を識別するための凡例（legend）を表示する役割を果たします。
# 凡例は通常、グラフの一角に配置され、各データ系列のラベルとそれに対応するマーカーや線のスタイルを示します。
# これにより、視覚的に各データ系列が何を表しているのかを容易に理解することができます。
plt.title('Stock Prices of Company A and Company B Over Time')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()

# Show the plot
plt.show()
```

## Exercise 5: Pie Chart
Represent the distribution of a company's market share using a pie chart.

labels = ['Company A', 'Company B', 'Company C', 'Company D'] 
market_share = [40, 30, 20, 10]

```python
import matplotlib.pyplot as plt

# First showing given data
labels = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [40, 30, 20, 10]

# パイチャートをつくる過程
plt.figure(figsize=(6, 4))
# autopct='%1.1f%%': これは各セグメントの中央に表示されるパーセンテージのフォーマットを指定します。
# %1.1f%%は浮動小数点数で1つの小数点を持つパーセンテージ形式です。
# startangle=90: これは円グラフが開始する角度を指定します。この例では、90度（垂直上）から開始します。
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'red', 'green', 'orange'])

# タイトルを載せる
plt.title('Market Share Distribution')

plt.show()
```

## Exercise 6: Linear equations
Solve the following system of linear equations graphically and algebraically:

y=3x−2

y=−2x+6

Steps:

Plot both equations on the same graph.

Find the point of intersection from the graph.

Use numpy's linalg.solve function to algebraically determine the exact intersection point (solution to the system of equations).

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions 関数を最初に定義する
def f1(x):
    return 3*x - 2

def f2(x):
    return -2*x + 6

# Generate x values
x = np.linspace(-3, 3, 400)

# Calculate y values
y1 = f1(x)
y2 = f2(x)

# Plotting both functions
plt.plot(x, y1, '-r', label='y=3x-2')
plt.plot(x, y2, '-b', label='y=-2x+6')
plt.title('Intersection of Two Linear Equations')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend(loc='upper right')
plt.show()
```
