````python
# My code for 1:

import numpy as np

array = np.array([[4, 6],[2, 8]])

print(f"This is the prepared array: \n{array}")

mean = np.mean(array)
print(f"The mean value of array is: {mean}") 

# My code for 2:

sum_column = np.sum(array, axis = 0)

sum_column_str = ' '.join(map(str, sum_column)) # map(str, sum_column): この部分はsum_column配列の各要素を文字列に変換します。join関数は、文字列の配列を一つの文字列に結合します。
print(f"The sum of each column is: [{sum_column_str}]")

# My code for 3:

minimum_value = np.min(array[1, :]) # array[1, :]は第2行のすべての要素を指す
print(f"The minimum value of the second row is: {minimum_value}")

# My code for 4:
# ブーリアン・インデックスを使用して、配列から5より大きい値をすべて除外する。

filtered_array = array[array > 5]
print(f"Values greater than 5: {filtered_array}")

# My code for 5:
# Multiply the entire matrix by 2 and subtract 3. (行列全体に2を掛け、3を引く。)

array = array * 2 - 3
print(f"The entire matrix multiplied by 2 and subtracted by 3: \n{array}")
````

````python
# Solution

# Import NumPy and create the array
import numpy as np

array = np.array([[4, 6], [2, 8]])
print("Original Array:")
print(array)
print("\n")

# 1: Calculate the mean of the entire array
mean_value = np.mean(array)
print(f"Mean of the entire array: {mean_value}")

# 2: Calculate the sum of each column
column_sum = np.sum(array, axis=0)  # axis=0 indicates sum along columns
print(f"Sum of each column: {column_sum}")

# 3: Find the minimum value of the second row
min_value = np.min(array[1, :])  # Indexing the second row and calculating the minimum
print(f"Minimum value of the second row: {min_value}")

# Task 4: Use boolean indexing to filter out all values greater than 5
filtered_values = array[array > 5]
print(f"Values greater than 5: {filtered_values}")

# Task 5: Multiply the matrix by 2 and subtract 3
transformed_array = (array * 2) - 3
print("Transformed Array:")
print(transformed_array)
````

## Exercise 7: Pandas DataFrame Statistical Analysis

You are given a dataset containing the exam scores of students in three subjects: Math, English, and Science.

Objective: Perform a statistical analysis on the dataset to derive insights.

````python
# My code

import pandas as pd
import numpy as np

np.random.seed(42)  # To ensure reproducibility
data = {
    'Math': np.random.randint(50, 100, 100),
    'English': np.random.randint(50, 100, 100),
    'Science': np.random.randint(50, 100, 100)
}
df = pd.DataFrame(data, index= [f'Student {i}' for i in range(100)])

print(df)
````

1. Calculate the mean and standard deviation for each subject.
2. Find out how many students scored above 90 in Math.
3. Determine which subject had the highest mean score.
4. Compute the correlation between the scores of Math and English.
5. Find the student (by index) who had the highest average score across all subjects.

````python
# My code for 1:

mean_score = df.mean(axis = 0)

# axis=0: 行方向（縦）に沿った操作。各列のすべての行に対して関数（平均、標準偏差など）が適用されます。
# axis=1: 列方向（横）に沿った操作。各行のすべての列に対して関数が適用されます。
# この場合、科目ごと（列ごと）の平均値と標準偏差を求めたいので、axis=0 を使用しています。
# この指定がないと、デフォルトの動作が適用され、その動作は関数やライブラリによって異なる場合があります。
# 明示的に axis=0 を設定することで、意図した動作が確実に行われるようになります。

standard_deviation = df.std(axis = 0) # standard deviation = 標準偏差

print(f"The mean score for each subject is: \n{mean_score}")
print(f"The standard deviation for each subject is: \n{standard_deviation}")
````

````python
# Solution 1: Mean and standard deviation for each subject
mean_scores = df.mean()
std_dev_scores = df.std()

print(f"Mean Scores:\n{mean_scores}\n")
print(f"Standard Deviation Scores:\n{std_dev_scores}\n")
````

````python
# My code for 2:
# Find out how many students scored above 90 in Math.

above_90_math = df[df["Math"] > 90]
print(f"The number of students who scored above 90 in Math is: {len(above_90_math)}")
````

````python
# Solution 2: Students who scored above 90 in Math
students_above_90_math = df[df['Math'] > 90].shape[0]
print(f"Number of students who scored above 90 in Math: {students_above_90_math}\n")
````

````python
# My code for 3:
# Determine which subject had the highest mean score.

# max() と idxmax() の違いは、max() が最大値自体を返すのに対して、
# idxmax() は最大値を持つインデックス（この場合は科目名）を返す点です。

highest_mean_subject = mean_score.idxmax()
print(f"The subject which has the highest mean score is: {highest_mean_subject}")
````

````python
# Solution 3: Subject with highest mean score
highest_mean_subject = mean_scores.idxmax()
print(f"The subject with the highest mean score is: {highest_mean_subject}\n")
````

````python
# My code for 4:
# Compute the correlation between the scores of Math and English.
# 数学と英語の得点の相関関係を計算する。

correlation_math_english = df["Math"].corr(df["English"])
print(f"The correlation between the scores of Math and English is: {correlation_math_english:.2f}")

# .2f 浮動小数点数を小数点以下2桁まで表示するための指定
````

````python
# Solution 4: Correlation between Math and English scores
correlation_math_english = df['Math'].corr(df['English'])
print(f"Correlation between Math and English scores: {correlation_math_english:.2f}\n")
````

````python
# My code for 5:
# Find the student (by index) who had the highest average score across all subjects.

highest_average_score = df.mean(axis = 1).idxmax()
print(f"The student who had the highest average score across all subjects is: {highest_average_score}")
````

## Exercise 9: Plot some simple functions

a. $y=2x$

b. $y=x^2$

c. $y=2^x$

````python
# My code for 1:
# Plot the following functions in the same figure. Plot for the x-values 0 to 5 using an interval of 0.1.

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1) # 0から5まで0.1刻みでxの値を生成する

# Define the functions
y1 = 2 * x
y2 = x ** 2
y3 = 2 ** x

# Create the figure and plot space
plt.figure(figsize=(8, 6))

# Plotting the functions
plt.plot(x, y1, label='y=2x', color='red')
plt.plot(x, y2, label='$y=x^2$', color='blue')
plt.plot(x, y3, label='$y=2^x$', color='green')

# Adding grid
plt.grid(True)

# Adding title
plt.title("Plotting Multiple Functions")

# Adding x-axis and y-axis labels
plt.xlabel("X-values")
plt.ylabel("Y-values")

# Adding legend to differentiate the functions
plt.legend() # 凡例を表示する

# Show plot
plt.show()

````

## Exercise 13: Supply and demand

In economics we often assume that the quantity demanded of a good is getting smaller when the price increases. At the same time, the supply will increase when the price increase and it is more profitable to sell the good. Here is an example of a mathematical model that describes this behaviour:

$(1)\;\;q_D = 40 - 2p$

$(2)\;\;q_S = 6p$ 

a. Plot equation (1) and (2) in the same figure. The $q$'s should be on the $y$-axis and the $p$ on the $x$-axis. Where does the two lines cross?

Hint: Plot the two functions from 0 to 10, e.g., `p = np.arange(0, 10, 0.1)`

b. The equations (1) and (2) is a system of linear equations if we set $q_D=q_S=q$. Rewrite the two equations on the form of $Ax=b$, where
$$x=\left[\begin{matrix}
 p \\
 q
\end{matrix}\right]$$
and solve this system using NumPy.

````python
# My code
````

````python
# Solution 13a:

# Let's create the p-values we will use when calculating the functions

p = np.arange(0, 10, 0.1)

# Then we can calculate the two functions:

qD = 40 - 2*p
qS = 6*p

# And we can plot them in the same figure:

plt.plot(p, qD, 
         lw=3, label='Demand curve')
plt.plot(p, qS, 
         lw=3, label='Supply curve')
plt.grid()
plt.legend()

print('The two lines cross when qD=qS=30 and p=5')
````

````python
# My code
````

````python
# Solution 13b:

# We can write the equations as follows 
# 2p + q = 40
# -6p + q = 0

# And then we have: 

A = np.array([[2, 1], 
              [-6, 1]])

b = np.array([[40], 
              [0]])

# And the solution to this system is:

np.linalg.solve(A, b)
````
