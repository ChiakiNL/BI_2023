```python
# In class exercise: Basic SELECT Query
# Write a SQL query to select all records from the Employees table.

import sqlite3
con = sqlite3.connect("auto_dealership_database.db")
cur = con.cursor()
res = cur.execute("SELECT * FROM Employees")
employees_table = res.fetchall() # res means result
print(employees_table)
```

```python
# In class exercise: List all employees
# Use the sqlite3 library and write a SQL query to list all employees' names and departments from the Employees table.

res = cur.execute("SELECT Name, Department FROM Employees")
names_departments = res.fetchall()
print(names_departments)
```
